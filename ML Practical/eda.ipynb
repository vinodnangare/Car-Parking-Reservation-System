```python
import pandas as pd
data_set = pd.read_csv('netflix.csv')
print(data_set.head())
```

```python
print(data_set.tail())
```

```python

from pandas_profiling import ProfileReport

profile = ProfileReport(data_set, title="Netflix data EDA Report", explorative=True)
report_path = 'netflix_report.html'
profile.to_file(report_path)
print(f"EDA report has been saved to {report_path}")
```

```python
import matplotlib.pyplot as plt

grouped_year=data_set.groupby('release_year')['show_id'].count().sort_index(ascending=False)
grouped_year.plot(kind='line', marker='o', color='orange')
plt.title("Number Of Shows By Year Of Release")
plt.xlabel("Year")
plt.ylabel("Count")
plt.grid(True)
plt.show()
```

```python

grouped_type = data_set.groupby('type')['show_id'].count()
grouped_type.plot(kind='bar', color='blue')
plt.title("Number Of Shows By Type")
plt.xlabel("Type")
plt.ylabel("Count")
plt.grid(True)
plt.show()
```
